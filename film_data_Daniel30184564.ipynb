```python
#фільтруваня не потрібних колумів
df=Fdata[['genres','movie_title','imdb_score','movie_imdb_link']].copy()
```

```python
#дісплею вмі жанри
genres_all_movies = [df.loc[i]['genres'].split('|') for i in df.index]
# сортую жанри в альфабетному порядку
genres = sorted(list(set([item for sublist in genres_all_movies for item in sublist])))
genres
```

```python
# ініціалізую лист з фільм_айді та назвами фільмів для подальшого вик
full_data = []
movie_titles = []
```

```python
# 1. Vectorized operation to strip whitespace from title and IMDb link columns
df['movie_title'] = df['movie_title'].str.strip()
df['movie_imdb_link'] = df['movie_imdb_link'].str.strip()

# 2. Define a function to process each row and return movie data
def process_row(row):
    movie_data = [1 if genre in row['genres'].split('|') else 0 for genre in genres]
    return movie_data

# 3. Apply the function to each row and store the results
full_data = df.apply(process_row, axis=1).tolist()

# 4. Extract movie titles and IMDb links
movie_titles = [(row['movie_title'], index, row['movie_imdb_link']) for index, row in df.iterrows()]
```
